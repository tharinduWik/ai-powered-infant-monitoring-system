```python
# Step 1: Install and import dependencies
import sys
import subprocess
import os
import numpy as np
import pandas as pd
import cv2
import urllib.request
import warnings
warnings.filterwarnings('ignore')

print(f"NumPy version: {np.__version__}")
print(f"OpenCV version: {cv2.__version__}")
print(f"Pandas version: {pd.__version__}")
print()
print("✓ All imports successful!")
```

```python
# Step 2: Initialize face detectors
print("Initializing face detectors...")

# Load Haar Cascade (built-in, always available)
face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)
print(f"✓ Haar Cascade loaded from: {face_cascade_path}")

# Try to download DNN model (optional, for better accuracy)
model_dir = "./cv_models"
os.makedirs(model_dir, exist_ok=True)

prototxt_path = os.path.join(model_dir, "deploy.prototxt.txt")
weights_path = os.path.join(model_dir, "res10_300x300_ssd_iter_140000.caffemodel")

net = None
use_dnn = False

if not os.path.exists(prototxt_path) or not os.path.exists(weights_path):
    print("\nDownloading DNN face detector model (~10 MB, one-time)...")
    try:
        prototxt_url = "https://raw.githubusercontent.com/opencv/opencv_3rdparty/dnn_samples_face_detector_20170830/res10_300x300_ssd_iter_140000.caffemodel.prototxt"
        weights_url = "https://raw.githubusercontent.com/opencv/opencv_3rdparty/dnn_samples_face_detector_20170830/res10_300x300_ssd_iter_140000.caffemodel"
        
        print("  Downloading prototxt...")
        urllib.request.urlretrieve(prototxt_url, prototxt_path)
        print("  Downloading weights...")
        urllib.request.urlretrieve(weights_url, weights_path)
        print("  ✓ Model files downloaded!")
    except Exception as e:
        print(f"  Download failed: {e}")
        print("  Will use Haar Cascade (reliable fallback)")
        prototxt_path = None
        weights_path = None

# Load DNN if available
if prototxt_path is not None and weights_path is not None and os.path.exists(prototxt_path) and os.path.exists(weights_path):
    try:
        net = cv2.dnn.readNetFromCaffe(prototxt_path, weights_path)
        use_dnn = True
        print("✓ DNN face detector loaded (high accuracy mode)")
    except:
        print("⚠ DNN loading failed, using Haar Cascade")
        net = None
        use_dnn = False
else:
    print("ℹ Using Haar Cascade detector (fast, reliable)")

print("\n✓ Face detectors ready!")
```

```python
# Step 3: Define face detection functions

def detect_faces_dnn(image):
    """Detect faces using OpenCV DNN."""
    if net is None:
        return []
    
    h, w = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 1.0, (300, 300), [104, 117, 123], False, False)
    net.setInput(blob)
    detections = net.forward()
    
    faces = []
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > 0.3:  # Lowered threshold to catch more faces
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x, y, x2, y2) = box.astype("int")
            x, y = max(0, x), max(0, y)
            x2, y2 = min(w, x2), min(h, y2)
            if x2 > x and y2 > y:
                faces.append((x, y, x2-x, y2-y))
    
    return faces

def detect_faces_haar(image):
    """Detect faces using Haar Cascade."""
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Lowered minNeighbors from 5 to 3 for better detection
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)
    return [(int(x), int(y), int(w), int(h)) for x, y, w, h in faces]

print("✓ Face detection functions defined")
print(f"  - DNN threshold: 0.3 (sensitive)")
print(f"  - Haar minNeighbors: 3 (sensitive)")
```

```python
# Step 4: Define feature extraction function

def extract_face_features(image, image_path=""):
    """Extract texture features from face region."""
    try:
        # Detect faces
        if use_dnn:
            faces = detect_faces_dnn(image)
        else:
            faces = detect_faces_haar(image)
        
        if len(faces) == 0:
            return None
        
        # Use largest face
        x, y, w, h = max(faces, key=lambda f: f[2]*f[3])
        
        # Validate bounds
        if x + w > image.shape[1] or y + h > image.shape[0]:
            return None
        
        face_region = image[y:y+h, x:x+w]
        
        if face_region.size == 0 or face_region.shape[0] < 10 or face_region.shape[1] < 10:
            return None
        
        # Convert to grayscale
        gray = cv2.cvtColor(face_region, cv2.COLOR_BGR2GRAY)
        gray = cv2.normalize(gray, None, 0, 255, cv2.NORM_MINMAX)
        
        # 1. Edge density (Sobel)
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        edges = np.sqrt(sobelx**2 + sobely**2)
        edge_density = np.mean(edges) / 255.0
        
        # 2. Contrast (std dev of intensity)
        contrast = np.std(gray) / 255.0
        
        # 3. Brightness (mean intensity)
        brightness = np.mean(gray) / 255.0
        
        # 4. Texture variance (Laplacian)
        laplacian = cv2.Laplacian(gray, cv2.CV_64F)
        texture_var = np.var(laplacian) / (255.0**2)
        
        # 5. Histogram entropy
        hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
        hist = hist.flatten() / hist.sum()
        entropy = -np.sum(hist[hist > 0] * np.log2(hist[hist > 0]))
        entropy = entropy / 8.0
        
        return {
            "Edge_Density": float(edge_density),
            "Contrast": float(contrast),
            "Brightness": float(brightness),
            "Texture_Variance": float(texture_var),
            "Entropy": float(entropy),
            "Face_Size": w * h
        }
    except:
        return None

print("✓ Feature extraction function defined")
```

```python
# Step 5: Process all images

DATASET_PATH = "../data/raw/img/Pain Level/"
DEBUG = False  # Set True for detailed per-image logs

data = []
stats = {
    "total_files": 0,
    "images_loaded": 0,
    "faces_detected": 0,
    "faces_extracted": 0,
    "errors": 0
}

print("Starting image processing...")
print(f"Dataset path: {os.path.abspath(DATASET_PATH)}")
print()

# Process each pain level folder
for folder_name in sorted(os.listdir(DATASET_PATH)):
    folder_full_path = os.path.join(DATASET_PATH, folder_name)
    
    if not os.path.isdir(folder_full_path):
        continue
    
    # Parse pain level
    try:
        pain_level = float(folder_name)
        if pain_level <= 2.0:
            label = 0
            label_str = "Calm"
        elif pain_level <= 4.5:
            label = 1
            label_str = "Discomfort"
        else:
            label = 2
            label_str = "Severe Pain"
    except ValueError:
        continue
    
    # Get image files
    img_files = [f for f in os.listdir(folder_full_path)
                 if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff'))]
    
    if len(img_files) == 0:
        continue
    
    print(f"Processing: Level {folder_name:>3} ({label_str:15}) - {len(img_files):3} images...", end="")
    
    folder_extracted = 0
    for img_file in img_files:
        img_path = os.path.join(folder_full_path, img_file)
        stats["total_files"] += 1
        
        # Load image
        image = cv2.imread(img_path)
        if image is None:
            stats["errors"] += 1
            if DEBUG:
                print(f"\n  ERROR loading {img_file}")
            continue
        
        stats["images_loaded"] += 1
        
        # Extract features
        features = extract_face_features(image, img_path)
        
        if features:
            stats["faces_extracted"] += 1
            folder_extracted += 1
            features["Label"] = label
            features["Label_String"] = label_str
            features["Filename"] = img_file
            features["Pain_Level"] = pain_level
            data.append(features)
    
    print(f" -> Extracted: {folder_extracted}/{len(img_files)}")

print()
print("="*70)
print("PROCESSING COMPLETE")
print("="*70)
print(f"Total files scanned:        {stats['total_files']}")
print(f"Images loaded:              {stats['images_loaded']}")
print(f"Faces extracted:            {stats['faces_extracted']}")
print(f"Load errors:                {stats['errors']}")
if stats['images_loaded'] > 0:
    print(f"Success rate:               {100*stats['faces_extracted']/stats['images_loaded']:.1f}%")
```

```python
# Step 6: Save results

# Create DataFrame
df = pd.DataFrame(data)

# Ensure output directory exists
output_path = "../data/processed/processed_pain_img_biomarkers1.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Save
df.to_csv(output_path, index=False)

print("\n" + "="*70)
print("RESULTS SAVED")
print("="*70)
print(f"File: {os.path.abspath(output_path)}")
print(f"Samples: {len(df)}")
print(f"Features: {df.shape[1]}")
print()
print("Class distribution:")
print(df['Label_String'].value_counts().sort_index())
print()
print("Feature statistics:")
print(df[['Edge_Density', 'Contrast', 'Brightness', 'Texture_Variance', 'Entropy']].describe())
print()
print("✓ Dataset ready for model training!")
```
